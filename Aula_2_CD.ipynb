```python
missingno.bar(df)
missingno.matrix(df)
```

```python
# Eliminando dados faltantes
df_new = df.dropna(axis = 0, subset = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'pH', 'sulphates'])
```

```python
missingno.bar(df_new)
missingno.matrix(df_new)
```

```python
# Criando um plot com scatterplot e histograma de todos os dados
figsize=(12, 8)

sns.set_theme(style="ticks", palette="pastel")
scater = sns.pairplot(df_new.iloc[:,:], diag_kind =  'hist')
```

```python
sns.set_theme(style="ticks", palette="pastel")

plt.figure(figsize = (10,6))
sns.boxplot(x="quality", y="alcohol",
            hue="type", palette=["w", "r"],
            data=df_new)
sns.despine(offset=10, trim=True)
```

- Os vinhos com maiores teores de álcool são **os de qualidade 7 e 8 de ambos os tipos**. Já os vinhos com menores teores de álcool são **os vinhos de qualidade 5 de ambos os tipos**.
- O vinho de **qualidade 9** se apresentam como sendo apenas do tipo **branco**.
- O vinho de **qualidade 5** apresentam o maior número de *outliers* superiores quando analisamos o teor alcólico desses vinhos.

```python
df_1 = df_new[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 
               'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']]
plt.figure(figsize=(14,16))

for i,cat in enumerate(df_1):
    plt.subplot(5,3,i+1)
    sns.set_theme(style="whitegrid")
    sns.boxplot(x=cat, data=df_new)
    plt.title("Boxplot {}".format(cat))
plt.tight_layout()
plt.show()
```

Percebe-se a presença de **muitos *outliers*** em todas as variáveis numéricas do banco. 
Para reduzir a presença desses *outliers*, será feita a utilização do teste estatístico *Z- scores*.

```python
# Utilizando o z- scores para eliminar outliers dos dados
from scipy import stats


z_scores= stats.zscore(df_1)
z_scores
```

```python
# Aplicaremos eliminação dos dados que obtiverem valores de z_scores > 3
abs_z_scores = abs(z_scores)
filtered_entries = (abs_z_scores < 3).all(axis=1)
df_new = df_new[filtered_entries]
df_new
```

```python
df_1 = df_new[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 
               'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']]
plt.figure(figsize=(14,16))

for i,cat in enumerate(df_1):
    plt.subplot(5,3,i+1)
    sns.set_theme(style="whitegrid")
    sns.boxplot(x=cat, data=df_new)
    plt.title("Boxplot {}".format(cat))
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (10,6))
sns.scatterplot(x = 'fixed acidity', y = 'pH', data = df_new, hue = "quality")
```

A grande concentração de amostras de vinhos está localizada na faixa de 5 - 10 de acidez, enquanto que o pH está entre 2,9 - 3,6 para as 7 qualidades  diferentes de vinhos analisadas.

```python
plt.figure(figsize = (8,6))
sns.countplot(df_new["quality"], order = df_new["quality"].value_counts().index)
```

```python
labels = df_new["type"].unique()
size = df_new["type"].value_counts().values

plt.figure(figsize = (8,8))
plt.pie(size, labels = labels, autopct = "%1.1f%%")
```

A grande **maioria das amostras** presente no banco de dados são da **qualidade 6, 5 e 7 respectivamente**.

Os **vinhos da qualidade 3 e 9**, são as qualidades de vinhos que **menos possuem observações no banco de dados**.

**78.7% dos vinhos são brancos e 21.3% são vermelhos ou tinto.**

```python
df_new["quality"].value_counts()
```

```python
#Gráfico de correlações entre as variáveis (Heatmap)
plt.figure(figsize=(15, 8))

corre = df_new.corr()
sns.heatmap(corre, annot= True, cmap= 'coolwarm', vmax=.9, center=0,
            square=True, linewidths=.5, fmt= '.1f', cbar_kws={"shrink": .5})
```

A partir da análise do gráfico de correlações podemos observar que:

- As variáveis: **free sulfur dioxide vs. total sulfur dioxide** e **density vs. residual sugar** possuem uma alta correlação positiva (= 0.7);
- As variáveis: **density vs. alcohol** possuem uma alta correlação negativa (= -0.7)
