```python
df_1 = df_new[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 
               'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']]
plt.figure(figsize=(14,16))

for i,cat in enumerate(df_1):
    plt.subplot(5,3,i+1)
    sns.set_theme(style="whitegrid")
    sns.boxplot(x=cat, data=df_new)
    plt.title("Boxplot {}".format(cat))
plt.tight_layout()
plt.show()
```

Percebe-se a presença de **muitos *outliers*** em todas as variáveis numéricas do banco. 
Para reduzir a presença desses *outliers*, será feita a utilização do teste estatístico *Z- scores*.

```python
# Utilizando o z- scores para eliminar outliers dos dados
from scipy import stats


z_scores= stats.zscore(df_1)
z_scores
```

```python
# Aplicaremos eliminação dos dados que obtiverem valores de z_scores > 3
abs_z_scores = abs(z_scores)
filtered_entries = (abs_z_scores < 3).all(axis=1)
df_new = df_new[filtered_entries]
df_new
```

```python
df_1 = df_new[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 
               'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']]
plt.figure(figsize=(14,16))

for i,cat in enumerate(df_1):
    plt.subplot(5,3,i+1)
    sns.set_theme(style="whitegrid")
    sns.boxplot(x=cat, data=df_new)
    plt.title("Boxplot {}".format(cat))
plt.tight_layout()
plt.show()
```

# **Criação de um modelo para predição da qualidade do vinho**

#### **Transformando a variável *type*** 

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

df_new['type'] = label_encoder.fit_transform(df_new['type'])
```

```python
df_new
```

#### **Divisão do banco de dados**

```python
from sklearn.model_selection import train_test_split
X = df_new.drop("quality", axis=1)
y = df_new['quality']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

#### **Padronização dos dados**

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

#### **Criando um modelo de RNA para classificação de multiclasses**

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.activations import linear, relu, sigmoid

tf.random.set_seed(1234) # for consistent results
model = Sequential(
    [               
        tf.keras.Input(shape=(12,)),
        Dense(25, activation = 'relu', name = "L1"),
        Dense(15, activation = 'relu', name = "L2"),
        Dense(10, activation = 'linear', name = "L3"),    
    ], name = "my_model" 
)
```

```python
model.summary()
```

```python
[layer1, layer2, layer3] = model.layers
W1,b1 = layer1.get_weights()
W2,b2 = layer2.get_weights()
W3,b3 = layer3.get_weights()
print(f"W1 shape = {W1.shape}, b1 shape = {b1.shape}")
print(f"W2 shape = {W2.shape}, b2 shape = {b2.shape}")
print(f"W3 shape = {W3.shape}, b3 shape = {b3.shape}")
```

```python
model.compile(
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), metrics = ['accuracy']
    )

history = model.fit(
    X_train,y_train, 
    epochs= 200
    )
```

```python
import numpy as np
p_nonpreferred = model.predict(X_test)
print(p_nonpreferred [:2])
print("largest value", np.max(p_nonpreferred), "smallest value", np.min(p_nonpreferred))
```

```python
 y_pred_test = []
for i in range(len(y_test)):
    print( f"{p_nonpreferred[i]}, category: {np.argmax(p_nonpreferred[i])}")
    y_pred_test.append(np.argmax(p_nonpreferred[i]))
```

```python
y_pred_test = np.array(y_pred_test)
```

```python
print(np.concatenate((y_pred_test.reshape(len(y_pred_test),1), y_test.values.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred_test)
print(cm)
accuracy_score(y_test, y_pred_test)
```
